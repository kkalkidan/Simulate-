```python

class Satellite:

    # Initializer / Instance Attributes
    def __init__(self, orbit_number, satellite_number, position, side):
        self._orbit_number = orbit_number
        self._satellite_number = satellite_number
        self._position = position
        self._side = side

    def Compare(self, s):
        return self.orbitNumber() == s.orbitNumber() and \
               self.satelliteNumber() == s.satelliteNumber()

    def Position(self):
        return self._position  # angle

    def changePosition(self, position):
        self._position = position

    def satelliteNumber(self):
        return self._satellite_number

    def orbitNumber(self):
        return self._orbit_number

    def side(self):  # 1, -1, 0
        return self._side

    def changeSide(self, side):
        self._side = side
```

```python
#initialization of the position of the satellite and side configurations 

satellite_list = [None]*66

side1 = [65.45, 32.72, 0, -32.72, -65.45]
side2 = [-81.81, -49.09, -16.36, 16.36, 49.09, 81.81]

for i in range(66): 
    if((i % 11) + 1 < 6):
        pos = side1[i%11]
        side = 1
    else:
        pos = side2[(i%11)-5]
        side = 2
        
    satellite_list[i] = Satellite(int(i/11) + 1, i%11 + 1, pos, side )
    
print("orbit number", "satellite Number", "position", "side")

for j in satellite_list:
    
    print(j.orbitNumber(), j.satelliteNumber(), j.Position(), j.side())
```

```python

class IridiumConnectivity(Satellite):


    def __init__(self, satellite_list):

#       self.satellite_list = satellite_list
        self.connectivity_graph = [[satellite_list[(j*11)+i] for i in range(11)] for j in range(6)]
        print(len(self.connectivity_graph))

    def Polar(self, s):

        return s.Position() > 70 or s.Position() < -70

    def Compare(self, s0, s1):

        return s0.satelliteNumber() - s1.satelliteNumber()

    def HigherSatNum(self, s):

        return self.connectivity_graph[s.orbitNumber()-1][(s.satelliteNumber())%11]

    def LowerSatNum(self, s):

        return self.connectivity_graph[s.orbitNumber()-1][(s.satelliteNumber()+9)%11]

    def HigherOrbitNum(self, s):

        if(s.orbitNumber() < 6 and s.Position() < 70 \
            and s.Position() > -70):
            return self.connectivity_graph[s.orbitNumber()][s.satelliteNumber()-1]

        return None

    def LowerOrbitNum(self, s):

        if(s.orbitNumber() > 1 and s.Position() < 70 \
            and s.Position() > -70):
            return self.connectivity_graph[s.orbitNumber()-2][s.satelliteNumber()-1]

        return None

    def getConnGraph(self):

        return self.connectivity_graph

    def Parallel(self, s0, s1):

        return s0.satelliteNumber() == s1.satelliteNumber()

    def moveAlongOrbit(self, s0, s1):

        comp = s0.satelliteNumber() - s1.satelliteNumber()
        if(comp > 0):
            return self.LowerSatNum(s0)
        else:
            return self.HigherSatNum(s0)

    def moveAcrossOrbit(self, s0, s1):

        comp = s0.orbitNumber() - s1.orbitNumber()
        if(comp > 0):
            return self.LowerOrbitNum(s0)
        else:
            return self.HigherOrbitNum(s0)


    def FindManhattan(self, s0, s1):

        count = 0
#         print("Source ", s0.orbitNumber(), s0.satelliteNumber(), self.Polar(s0))
#         print("Destination ", s1.orbitNumber(), s1.satelliteNumber(), self.Polar(s1))

        
        while(not(self.Parallel(s0, s1)) and not(self.Polar(s1))):

            s0 = self.moveAlongOrbit(s0, s1)
#             print("Next hop ", s0.orbitNumber(), s0.satelliteNumber())
            count+=1
            

        while(self.Polar(s0) and not(s0.Compare(s1))):
            s0 = self.moveAlongOrbit(s0, s1)
#             print("Next hop ", s0.orbitNumber(), s0.satelliteNumber())
            count+=1
              
        while(not(s0.Compare(s1))):     
            if(s0.orbitNumber() != s1.orbitNumber()):
                s0 = self.moveAcrossOrbit(s0, s1)
                
            else:
                s0 = self.moveAlongOrbit(s0, s1)
                
#             print("Next hop ", s0.orbitNumber(), s0.satelliteNumber())
            count+=1

        return count
```
